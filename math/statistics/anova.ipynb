# 方差分析
* 研究分类型自变量对数量型因变量的影响时，使用方差分析(analysis of variance,anova)。
* 它是用来对比因变量在不同组中的平均值的统计方法。
* 方差分析与回归分析的主要区别：方差分析中沿水平轴的自变量是分类变量，而回归分析中沿水平轴的自变量是数量变量。

## 方差分析公式
方差分析：假设数据按表11.7的形式排列——每一组(类)观测值占一列：

![](http://ou8qjsj0m.bkt.clouddn.com//17-10-26/91662940.jpg)

* 分类自变量的平方和： $\sum{n_i(\bar{y_i}-\bar{y})^2}$
* 残差平方和： $\sum\sum{(\bar{y_{ij}}-\bar{y_i})^2}$
* 总平方和： $\sum\sum(\bar{y_{ij}}-\bar{y})^2$
* ![](http://ou8qjsj0m.bkt.clouddn.com//17-10-26/81677578.jpg)

公式中的k是组数(自变量的值的个数)，n是数据中的总观测数：

![](http://ou8qjsj0m.bkt.clouddn.com//17-10-26/21694944.jpg)

* 平方和的自由度
    * 分类变量的自由度=k-1
    * 残差的自由度=n-k
    * 总自由度=n-1

## 假设检验F检验
零假设：在暴力犯罪率的问题中零假设描述的是7个地区的犯罪率相等，即H0:μ1=μ2=μ3=μ4=μ5=μ6=μ7

F变量:

方差分析表提供了包括用来计算R和F-值的数据。用自由度对自变量和残差变量的效应先做一个调整然后再进行对比就可以得到F的值了。通常方差分析还包括F的p-值。

![](http://ou8qjsj0m.bkt.clouddn.com//17-10-26/35294956.jpg)

F的值就是为了比较自变量和残差变量的效应。在计算中我们对比的不是平方和而是平方和除以它们各自的自由度以得到每个自由度上的平方和。这个量称为均方(mean square)，之所以这样是因为它代表的是效应在自由度上的平均。

得到F值为2.11(自由度为6和41的F变量的观测值)。F≥2.11的概率是0.072。因此在7组48个观测的情况下得到$R^2$大于等于0.24的概率是边缘统计显著的。

## 配对分析
如果在同一个实验单元上进行重复测量，就会得到在每个单元中有两个观测值的数据。而处理这类数据时所用的分析方法叫做配对分析。 表11.6提供了48个大洲爆率犯罪率的配对数据。它是由已经在表1.1中给出的1986年数据和1992年数据组成。

### t检验
为了找出两个时间点上的数据是否有区别，我们求平均。差异的均值等于119；这说明暴力犯罪率还是有一些变化的。为判断这一变换是否只是偶然因素引起的，我们
1. 建立零假设：差异的总均值是0。
2. 我们利用检验单个均值的t-检验。对于该数据，t值是8.74自由度为47，与之相应的p-值小于0.0001，现在我们有了拒绝零假设的一个有力证据。
3. 公式
    1. 先从第i个观测单元中计算两次观测的差：di=yBi-yAi
    2. 其中yBi是第一次的观测值(起始)，yAi是第二次的观测值(结束)。然后求di的均值$\bar{d}$以及标准差s，最后转成t-变量： $t=\frac{\bar{d}}{s/\sqrt{n}}$, d.f. n-1
    3. 在前面的例子中上式为：$t=\frac{118.47}{93.93/\sqrt{48}}=8.74$ , d.f. 47 

### 符号检验z检验
与t检验不同，符号检验主要考虑的不是两次观测时数值改变了多少而是数值变化的趋势：如果实际上两次观测中数值没有什么变化，则它们的差值就只是在随机变化，没有规律。在我们的例子中48个差值中，只有4个负数，而正数有44个之多。我们利用二项分布来检验“差为正的概率是0.5”的零假设。二项分布表中不包括48个观测值那么多情况，我们通过计算z来得到p-值。这里z=6.21，得到p-值小于0.0001。因此我们拒绝犯罪率没有变化的零假设。
