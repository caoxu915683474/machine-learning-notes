# 两个顺序变量的秩方法
## 顺序变量
顺序变量(rank variable)，是以数量(或多或少)术语来比较个体的某些特征的变量。例如，选民可能对选举的结果非常感兴趣、有点兴趣或不感兴趣。社会地位可以被分为上层阶级、中产阶级和下层阶级。对于这些顺序变量，我们用词(word)来表示变量的值。另外一类顺序变量，如官员年收入从1到10进行排序，10是最高的。这类顺序变量，我们用数(number)来代表变量的值。

### t检验
计算t-检验的统计量： $t=\frac{r_s}{\sqrt{1-r_s^2}}\sqrt{n-2}=\frac{-0.09}{\sqrt{1-(-0.09)^2}}\sqrt{6-2}=0.17,df=4$

提出零假设，认为两个变量的关系仅处于偶然。把rs转化为相应的一个t-变量，自由度为n-2=6-2=4，t=-0.17。r≤-0.09的概率为0.44。这个p-值是不显著的。因此可以说这两个变量的关系仅出于偶然。
