## 指数函数
$y=a^x,(a>0,a \ne 1)$

- 定义域：$(-\infty,+\infty)$
- 值域：$(0,+\infty)$

图形过（0，1）点，a>1时，单调增加；a<时，单调减少。

![](http://ou8qjsj0m.bkt.clouddn.com//18-1-5/19062310.jpg)

常用公式：

- $a^{-x}=\frac{1}{a^x}$
- $a^{\frac{m}{n}}=\sqrt[n]{a^m},a>0$
- $a^m \cdot a^n = a^{m+n}$
- $\frac{a^m}{a^n}=a^{m-n}$
- $(a^m)^n=a^{mn}$
- $(ab)^n=a^nb^n$

## 对数函数
$y=log_a x,(a>0,a \ne 1)$

- 定义域：$(0,+\infty)$
- 值域：$(-\infty,+\infty)$

![](http://ou8qjsj0m.bkt.clouddn.com//18-1-5/42630338.jpg)

与指数函数互为反函数，图形过（1，0）点，a>1时，单调增加；a<1时，单调减少
。

常用公式：

- $log_{10}x=lgx,log_ex=Inx$
- $log_amn=log_am+log_an$
- $log_a\frac{m}{n}=log_am-log_an$
- $log_am^n=nlog_am$
- $log_ab=\frac{log_cb}{log_ca}$
- $log_a1=0$
- $log_aa=1$
- $a^x=y \Leftrightarrow x=log_ay,a^{log_aN}=N,y=e^{Iny}$

## 三角函数
$y=sinx,(-\infty,+\infty)$,奇函数、有界函数、周期函数 $(2\pi)$；

$y=cosx,(-\infty,+\infty)$,偶函数、有界函数、周期函数 $(2\pi)$；

![](http://ou8qjsj0m.bkt.clouddn.com//18-1-5/46125484.jpg)

$y=tgx=\frac{sinx}{cosx}, x \ne k\pi+\frac{\pi}{2},k=0,\pm1,\pm2,\cdots$的一切实数，奇函数、周期函数 $(\pi)$;

$y=ctgx=\frac{cosx}{sinx}, x \ne k\pi,k=0,\pm1,\pm2,\cdots$的一切实数，奇函数、周期函数 $(\pi)$;

![](http://ou8qjsj0m.bkt.clouddn.com//18-1-5/95876912.jpg)

$y=secx=\frac{1}{cosx},y=cscx=\frac{1}{sinx}$
